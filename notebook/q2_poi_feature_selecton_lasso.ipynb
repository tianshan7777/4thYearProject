```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LassoCV, Lasso
```

```python
data = pd.read_csv('rent_clean.csv')
```

```python
#tourism
test = data[['price_per_sqm','number_of_whole_rooms', 'floor', 'number_of_half_rooms', 'furnish_furnished', 'view_garden', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'heating_central_heating', 'heating_electric', 'heating_circulating', 'heating_district', 'available_immediately','tourism_apartment','tourism_attraction','tourism_gallery','tourism_guest_house','tourism_hostel','tourism_hotel','tourism_information','tourism_motel','tourism_museum']]
test = test.dropna()

X = test[['number_of_whole_rooms', 'floor', 'number_of_half_rooms', 'furnish_furnished', 'view_garden', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'heating_central_heating', 'heating_electric', 'heating_circulating', 'heating_district', 'available_immediately','tourism_apartment','tourism_attraction','tourism_gallery','tourism_guest_house','tourism_hostel','tourism_hotel','tourism_information','tourism_motel','tourism_museum']]
y = test['price_per_sqm']

reg = LassoCV()
reg.fit(X, y)
print('Best alpha using built-in LassoCV:', reg.alpha_)
print('Best score using built-in LassoCV:', reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.grid(True)
plt.title("Feature importance using Lasso Model")
```
