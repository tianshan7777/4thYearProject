```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
```

```python
data = pd.read_csv('kmeans_2.csv')
set_0 = data[data.cluster == 0]
set_1 = data[data.cluster == 1]
set_2 = data[data.cluster == 2]
set_3 = data[data.cluster == 3] 
set_4 = data[data.cluster == 4]
set_5 = data[data.cluster == 5]
set_6 = data[data.cluster == 6]
set_7 = data[data.cluster == 7]
set_8 = data[data.cluster == 8]

set_0.head()
```

```python
#set 0
X = set_0[['number_of_whole_rooms', 'floor', 'number_of_half_rooms', 'furnish_furnished', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'utility_double_comfort', 'heating_gas', 'heating_wall_heating', 'heating_central_heating', 'heating_electric', 'heating_floor', 'heating_circulating', 'heating_district', 'condition_of_property_renovated']]
y = set_0[['price_per_sqm']]

scores = []
cv = KFold(n_splits = 10, random_state = 42, shuffle = False)
for train_index, test_index in cv.split(X):
    X_train, X_test, y_train, y_test = X.iloc[train_index.tolist()], X.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    my_model = LinearRegression().fit(X_train, y_train)
    scores.append(my_model.score(X_test, y_test))

r_square = np.mean(scores)
adj_r = 1-(1-r_square)*(len(y)-1)/(len(y)-X.shape[1]-1)

print(r_square)
print(adj_r)
        
```
