```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#input data
data = pd.read_csv('rent_origin.csv')
```

```python
data['price_per_sqm'].describe()
```

```python
plt.figure(figsize=(10,10))
plt.hist(data['district'], bins=23, alpha=0.5)
plt.show()
```

```python
sns.catplot(x = 'district', y = 'price_per_sqm', data = data)
```

```python
plt.figure(figsize=(10,10))
plt.hist(data['number_of_whole_rooms'], bins=30, alpha=0.5)
plt.show()
```

```python
plt.scatter(data['number_of_whole_rooms'], data['price_per_sqm'], color = 'green')
plt.xlabel('number_of_whole_rooms', fontsize = 14)
plt.ylabel('price_per_sqm', fontsize = 14)
plt.grid(True)
plt.show()
```

```python
sns.countplot(x = "type_of_the_building", data = data)
```
