```python
test = data[['price_per_sqm','apartment','number_of_whole_rooms','number_of_half_rooms','floor','furnish_furnished','furnish_partly_furnished','view_garden','view_panoramic','view_street','balcony','parking','lift','air_conditioner','utility_double_comfort','utility_full_comfort','utility_luxury','heating_gas','heating_wall_heating','heating_combination','heating_ceiling','heating_central_heating','heating_electric','heating_floor','heating_circulating','heating_district','available_immediately','condition_of_property_renovated','condition_of_property_newly_build','condition_of_property_in_good_condition','condition_of_property_in_medium_condition','condition_of_property_needs_renovation','condition_of_property_mint','floors_of_the_building','energy_performance_certificate','ceiling_height','bathroom_separate','bathroom_together','orientation_north','orientation_south','orientation_east','orientation_west','orientation_northeast','orientation_northwest','orientation_southwest','orientation_southeast','garden_AC','attic_loft_conversion','attic_penthouse','attic_topfloor']]
test = test.dropna()

X = test[['apartment','number_of_whole_rooms','number_of_half_rooms','floor','furnish_furnished','furnish_partly_furnished','view_garden','view_panoramic','view_street','balcony','parking','lift','air_conditioner','utility_double_comfort','utility_full_comfort','utility_luxury','heating_gas','heating_wall_heating','heating_combination','heating_ceiling','heating_central_heating','heating_electric','heating_floor','heating_circulating','heating_district','available_immediately','condition_of_property_renovated','condition_of_property_newly_build','condition_of_property_in_good_condition','condition_of_property_in_medium_condition','condition_of_property_needs_renovation','condition_of_property_mint','floors_of_the_building','energy_performance_certificate','ceiling_height','bathroom_separate','bathroom_together','orientation_north','orientation_south','orientation_east','orientation_west','orientation_northeast','orientation_northwest','orientation_southwest','orientation_southeast','garden_AC','attic_loft_conversion','attic_penthouse','attic_topfloor']]
y = test[['price_per_sqm']]

reg_aic = LassoLarsIC(criterion='aic')
reg_aic.fit(X, y)
print('Best alpha using built-in Lasso AIC:', reg_aic.alpha_)
print('Best score using built-in Lasso AIC:', reg_aic.score(X,y))
coef = pd.Series(reg_aic.coef_, index = X.columns)
                 
```

```python
imp_coef = coef.sort_values()
import matplotlib
plt.figure(figsize=(30,20))
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.grid(True)
plt.savefig('lasso_AIC_ingatlan_extra.png')
plt.title("Feature importance using Lasso AIC Model")
```
