```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
```

```python
data = pd.read_csv('rent_clean.csv')
```

```python
test = data[['price_per_sqm', 'number_of_whole_rooms', 'number_of_half_rooms', 'furnish_furnished',  'view_garden', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'heating_gas', 'heating_wall_heating', 'heating_central_heating', 'heating_electric', 'heating_floor', 'heating_circulating', 'heating_district', 'available_immediately',  'parking', 'price_per_sqm', 'number_of_whole_rooms', 'number_of_half_rooms', 'furnish_furnished',  'view_garden', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'heating_gas', 'heating_wall_heating', 'heating_central_heating', 'heating_electric', 'heating_floor', 'heating_circulating', 'heating_district', 'available_immediately', 'parking','distace_to_danube','quadratic_danube','distace_to_city_centre','quadratic_centre','distace_to_keleti','quadratic_keleti','distace_to_nyugati','quadratic_nyugati','distance_to_deli','quadratic_deli','distance_to_airport','quadratic_airport','distance_to_uni1', 'quadratic_uni1','distance_to_uni5','quadratic_uni5','distance_to_uni13','quadratic_uni13','distance_to_uni16','quadratic_uni16']]
test = test.dropna()

X = test[['number_of_whole_rooms', 'number_of_half_rooms', 'furnish_furnished',  'view_garden', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'heating_gas', 'heating_wall_heating', 'heating_central_heating', 'heating_electric', 'heating_floor', 'heating_circulating', 'heating_district', 'available_immediately',  'parking', 'price_per_sqm', 'number_of_whole_rooms', 'number_of_half_rooms', 'furnish_furnished',  'view_garden', 'view_panoramic', 'view_street', 'balcony', 'lift', 'air_conditioner', 'heating_gas', 'heating_wall_heating', 'heating_central_heating', 'heating_electric', 'heating_floor', 'heating_circulating', 'heating_district', 'available_immediately', 'parking','distace_to_danube','quadratic_danube','distace_to_city_centre','quadratic_centre','distace_to_keleti','quadratic_keleti','distace_to_nyugati','quadratic_nyugati','distance_to_deli','quadratic_deli','distance_to_airport','quadratic_airport','distance_to_uni1', 'quadratic_uni1','distance_to_uni5','quadratic_uni5','distance_to_uni13','quadratic_uni13','distance_to_uni16','quadratic_uni16']]
y = test[['price_per_sqm']]

reg = LassoCV()
reg.fit(X, y)
print('Best alpha using built-in LassoCV:', reg.alpha_)
print('Best score using built-in LassoCV:', reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.grid(True)
plt.title("Feature importance using Lasso Model")
```
